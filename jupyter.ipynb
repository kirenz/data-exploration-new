# Barplot

```python
import pandas as pd
import altair as alt
```

## Datenimport

```python

ROOT = "https://raw.githubusercontent.com/kirenz/datasets/master/"
DATA = "loans.csv"
df = pd.read_csv(ROOT + DATA)
```

```python
alt.Chart(df).mark_bar().encode(
    x='homeownership',
    y='count(homeownership)'
)
```
